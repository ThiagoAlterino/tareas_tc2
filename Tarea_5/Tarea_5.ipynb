### 1) Obtener los parametros Z, Y o ABCD

#### Red T
![RedT](./archivos/RedT.png)

Ecuaciones que definen al cuadripolo utilizando parametros Z: $V_1 = Z_{11}I_1 + Z_{12}I_2 \quad y \quad V_2 = Z_{21}I_1 + Z_{22}I_2$

Primero se calcula los parametros $Z_{11}$ y $Z_{12}$, para esto hay que hacer $I_2 = 0$ y se inyecta una corriente $I_1$.
Circuito representativo:
![Parametros_Z11_Z21](./archivos/Parametros_Z11_Z21.png)

$$V_1 = Z_A I_1 + Z_B I_1 \quad \implies \quad V_1 = (Z_A + Z_B) I_1 \quad \implies \quad \boxed {Z_{11} = \frac{V_1}{I_1} = Z_A + Z_B}$$
$$V_B = Z_B I_1 \quad y \quad V_B = V_2 \quad \implies \quad V_2 = Z_B I_1 \quad \implies \quad \boxed{Z_{21} = \frac{V_2}{I_1} = Z_B}$$

Luego, para calcular los parametros $Z_{12}$ y $Z_{22}$ hay que hacer $I_1 = 0$ y se inyecta una corriente $I_2$
![Parametros_Z12_Z22](./archivos/Parametros_Z12_Z22.png)

$$V_1 = Z_B I_2 \quad \implies \quad \boxed{Z_{12} = \frac{V_1}{I_2} = Z_B}$$
$$V_2 = Z_B I_2 + Z_C I_2 \quad \implies \quad V_2 = (Z_B + Z_C)I_2 \quad \implies \quad \boxed{Z_{22} = \frac{V_2}{I_2} = Z_B + Z_C} $$

Quedando la matriz de parametros Z como:

$$
\begin{bmatrix}
V_1 \\
V_2
\end{bmatrix}
=
\begin{bmatrix}
Z_A + Z_B & Z_B \\
Z_B & Z_B + Z_C
\end{bmatrix}
\begin{bmatrix}
I_1 \\
I_2
\end{bmatrix}
$$

#### Red PI

Ecuaciones que definen al cuadripolo utilizando parametros Y: $I_1 = V_1 Y_{11} + V_2 Y_{12}\quad y \quad I_2 = V_1 Y_{21} + V_2 Y_{22}$

Para calcular $Y_{11}$ y $Y_{21}$ hay que cortocircuitar $V_2$ ($V_2 = 0$) y se exita con una fuente de tension $V_1$
Circuito representativo:

$$I_1 + I_2 = I_A \quad y \quad I_A = V_1 Y_A \quad ; \quad I_2 = -V_1 Y_C \quad \implies \boxed{Y_{21} = \frac{I_2}{V_1} = -Y_C} \quad y \quad I_1 = I_A - I_2 = V_1 Y_A - (-V_1 Y_C) = (Y_A + Y_C)V_1 \quad \implies \boxed{Y_{11} = \frac{I_1}{V_1} = Y_A + Y_C}$$

Para calcular los parametros $Y_{12}$ y $Y_{22}$ hay que cortocircuitar $V_1$ ($V_1 = 0$) y se exita con una fuente de tension $V_2$
Circuito representativo:

$$I_1 + I_2 = I_B \quad y \quad I_B = V_2 Y_B \quad ; \quad I_1 = -V_2 Y_C \quad \implies \boxed{Y_{12} = \frac{I_1}{V_2} = -Y_C} \quad y \quad I_2 = I_B - I_1 = V_2 Y_B + V_2 Y_C = (Y_B + Y_C)V_2 \quad \implies \boxed{Y_{22} = \frac{I_2}{V_2} = Y_B + Y_C}$$

Matriz de parametros Y:
$$
\begin{bmatrix}
I_1 \\
I_2
\end{bmatrix}
=
\begin{bmatrix}
Y_A + Y_C & -Y_C \\
-Y_C & Y_B + Y_C
\end{bmatrix}
\begin{bmatrix}
V_1 \\
V_2
\end{bmatrix}
$$

#### Red en puente simetrica

Ecuaciones que definen al cuadripolo utilizando parametros Z: $V_1 = Z_{11}I_1 + Z_{12}I_2 \quad y \quad V_2 = Z_{21}I_1 + Z_{22}I_2$

Primero se calcula los parametros $Z_{11}$ y $Z_{12}$, para esto hay que hacer $I_2 = 0$ y se inyecta una corriente $I_1$.
Circuito representativo:

$$\begin{align*}
I_1 &= I_A + I_B \quad y \quad I_A = I_B = \frac{I_1}{2} = \frac{V_1}{Z_A + Z_B} \quad \implies \quad I_1 = \frac{V_1}{Z_A + Z_B} + \frac{V_1}{Z_A + Z_B} = \frac{2}{Z_A + Z_B}V_1 \quad \implies \quad \boxed{Z_{11} = \frac{V_1}{I_1} = \frac{Z_A + Z_B}{2}} \\ \\
M_1 \quad : \quad 0 &= V_B - V_A - V_2 \quad \implies \quad V_2 = V_B - V_A = Z_B I_B - Z_A I_A = Z_B \frac{I_1}{2} - Z_A \frac{I_1}{2} = \frac{(Z_B - Z_A)}{2}I_2 \quad \implies \quad \boxed{ Z_{21} = \frac{V_2}{I_1} = \frac{(Z_B - Z_A)}{2}}
\end{align*}$$

Luego, para calcular los parametros $Z_{12}$ y $Z_{22}$ hay que hacer $I_1 = 0$ y se inyecta una corriente $I_2$
$$\begin{align*}
I_2 &= I_A + I_B \quad y \quad I_A = I_B = \frac{I_2}{2} = \frac{V_2}{Z_A + Z_B} \quad \implies \quad I_2 = \frac{2}{Z_A + Z_B} V_2 \quad \implies \quad \boxed{Z_{22} = \frac{V_2}{I_2} = \frac{Z_A + Z_B}{2}} \\ \\
M_1 \quad : \quad 0 &= V_B - V_A - V_1 \quad \implies \quad V_1 = Z_B I_B - Z_A I_A = Z_B \frac{I_2}{2} - Z_A \frac{I_2}{2} = \frac{(Z_B - Z_B)}{2}I_2 \quad \implies \quad \boxed{Z_{12} = \frac{V_1}{I_2} = \frac{(Z_B - Z_B)}{2}}
\end{align*}$$

Quedando la matriz de parametros Z como:

$$
\begin{bmatrix}
V_1 \\
V_2
\end{bmatrix}
=
\begin{bmatrix}
\frac{Z_A + Z_B}{2} & \frac{Z_B - Z_A}{2} \\
\frac{Z_B - Z_A}{2} & \frac{Z_A + Z_B}{2}
\end{bmatrix}
\begin{bmatrix}
I_1 \\
I_2
\end{bmatrix}
$$

### 2) Aplicar transformación de parámetros para hallar los Z, Y si lo obtenido fue ABCD o viceversa

Transformacion para obtener la matriz Y de la red T usando la matriz de parametros Z:
$$Y = \frac{adj(Z)}{\Delta Z} \quad y \quad \Delta Z = (Z_A + Z_B)(Z_B + Z_C)-Z_B² = Z_A Z_B + Z_A Z_C + Z_B² + Z_B Z_C - Z_B² \quad \implies \quad \Delta Z = Z_A Z_B + Z_A Z_C + Z_B Z_C$$
$$Y = \frac{1}{\Delta Z}
\begin{bmatrix}
Z_B + Z_C & -Z_B \\
-Z_B & Z_A + Z_B
\end{bmatrix}
= \begin{bmatrix}
\frac{Z_B + Z_C}{Z_A Z_B + Z_A Z_C + Z_B Z_C} & \frac{-Z_B}{Z_A Z_B + Z_A Z_C + Z_B Z_C} \\
\frac{-Z_B}{Z_A Z_B + Z_A Z_C + Z_B Z_C} & \frac{Z_A + Z_B}{Z_A Z_B + Z_A Z_C + Z_B Z_C}
\end{bmatrix}
$$

Transformacion para obtener la matriz ABCD de la red T usando la matriz de parametros Z:
$$A = \frac{Z_{11}}{Z_{21}} \quad ; \quad B = \frac{\Delta Z}{Z_{21}} \quad ; \quad C = \frac{1}{Z_{21}} \quad ; \quad D = \frac{Z_{22}}{Z_{21}} \quad \implies \quad \boxed{A = \frac{Z_A + Z_B}{Z_B}} \quad ; \quad \boxed{ B = \frac{Z_A Z_B + Z_A Z_C + Z_B Z_C}{Z_B}} \quad ; \quad \boxed{C = \frac{1}{Z_B}} \quad ; \quad \boxed{D = \frac{Z_B + Z_C}{Z_B}}$$

$$
\begin{bmatrix}
A & B \\
C & D
\end{bmatrix}
= \begin{bmatrix}
\frac{Z_A + Z_B}{Z_B} & \frac{Z_A Z_B + Z_A Z_C + Z_B Z_C}{Z_B} \\
\frac{1}{Z_B} & \frac{Z_B + Z_C}{Z_B}
\end{bmatrix}
$$

Transformacion para obtener la matriz Z de la red PI usando la matriz de parametros Y:
$$Z = \frac{adj(Y)}{\Delta Y} \quad y \quad \Delta Y = (Y_B + Y_C)(Y_A + Y_C) - Y_C² = Y_B Y_A + Y_B Y_C + Y_C Y_A$$
$$Z = \frac{1}{\Delta}
\begin{bmatrix}
Y_B + Y_C & Y_C \\
Y_C & Y_A + Y_C
\end{bmatrix}
= \begin{bmatrix}
\frac{Y_B + Y_C}{Y_B Y_A + Y_B Y_C + Y_C Y_A} & \frac{Y_C}{Y_B Y_A + Y_B Y_C + Y_C Y_A} \\
\frac{Y_C}{Y_B Y_A + Y_B Y_C + Y_C Y_A} & \frac{Y_A + Y_C}{Y_B Y_A + Y_B Y_C + Y_C Y_A}
\end{bmatrix}
$$

Transformacion para obtener la matriz ABCD de la red PI usando la matriz de parametros Y:

$$A = -\frac{Y_{22}}{Y_{21}} \quad ; \quad B = -\frac{1}{Y_{21}} \quad ; \quad C = -\frac{\Delta Y}{Y_{21}} \quad ; \quad D = -\frac{Y_{11}}{Y_{21}} \quad \implies \boxed{A = \frac{Y_A + Y_C}{Y_C}} \quad ; \quad \boxed{ B = \frac{1}{Y_C}} \quad ; \quad \boxed{C = \frac{Y_B Y_A + Y_B Y_C + Y_C Y_A}{Y_C}} \quad ; \quad \boxed{D = \frac{Y_A + Y_C}{Y_C}}$$
$$
\begin{bmatrix}
A & B \\
C & D
\end{bmatrix}
= \begin{bmatrix}
\frac{Y_A + Y_C}{Y_C} & \frac{1}{Y_C} \\
\frac{Y_B Y_A + Y_B Y_C + Y_C Y_A}{Y_C} & \frac{Y_A + Y_C}{Y_C}
\end{bmatrix}
$$

Transformacion para obtener la matriz Y de la red puente simetrica usando la matriz de parametros Z:
$$Y = \frac{adj(Z)}{\Delta Z} \quad y \quad \Delta Z = Z_A Z_B$$
$$Y = \frac{1}{\Delta Z}
\begin{bmatrix}
\frac{Z_A + Z_B}{2} & -\frac{Z_B - Z_A}{2} \\
-\frac{Z_B - Z_A}{2} & \frac{Z_A + Z_B}{2}
\end{bmatrix}
=\begin{bmatrix}
\frac{Z_A + Z_B}{2 Z_A Z_B} & -\frac{Z_B - Z_A}{2 Z_A Z_B} \\
-\frac{Z_B - Z_A}{2 Z_A Z_B} & \frac{Z_A + Z_B}{2 Z_A Z_B}
\end{bmatrix} 
$$

Transformacion para obtener la matriz ABCD de la red puente simetrica usando la matriz de parametros Z:
$$A = \frac{Z_{11}}{Z_{21}} \quad ; \quad B = \frac{\Delta Z}{Z_{21}} \quad ; \quad C = \frac{1}{Z_{21}} \quad ; \quad D = \frac{Z_{22}}{Z_{21}} \quad \implies \quad \boxed{A = \frac{Z_A + Z_B}{Z_B - Z_A}} \quad ; \quad \boxed{ B = \frac{2 Z_A Z_B}{Z_B - Z_A}} \quad ; \quad \boxed{C = \frac{2}{Z_B - Z_A}} \quad ; \quad \boxed{D = \frac{Z_A + Z_B}{Z_B - Z_A}}$$

$$
\begin{bmatrix}
A & B \\
C & D
\end{bmatrix}
= \begin{bmatrix}
\frac{Z_A + Z_B}{Z_B - Z_A} & \frac{2 Z_A Z_B}{Z_B - Z_A} \\
\frac{2}{Z_B - Z_A} & \frac{Z_A + Z_B}{Z_B - Z_A}
\end{bmatrix}
$$
