## 3)Simule la transferencia desnormalizada en Python. Se sugiere programar la transferencia normalizada y definir las normas de frecuencia e impedancia para desnormalizar. Verificar los requerimientos pedidos en 2)

$$\begin{align*}
Tipicamente\,para\,un\,pasabanda:T(S) &= \frac{K\frac{\omega_{0}}{Q}S}{S^{2}+\frac{\omega_{0}}{Q}S+\omega_{0}^{2}} \quad y \quad T(S) = \frac{S\frac{R_{4}+R_{5}}{RR_{5}C}}{S^{2}+\frac{1}{RC}S+\frac{R_{4}}{R_{1}R_{3}R_{5}CC_{2}}} \quad y \quad R_{1} = R_{3} = R_{4} = R_{5} \quad y \quad C = C_{2}
\end{align*}$$

$$\begin{align*}
T(S) &= \frac{S\frac{R_{1}+R_{1}}{RR_{1}C}}{S^{2}+\frac{1}{RC}S+\frac{R_{1}}{R_{1}R_{1}R_{1}CC}} \quad \implies \quad T(S) = \frac{S\frac{2R_{1}}{RR_{1}C}}{S^{2}+\frac{1}{RC}S+\frac{R_{1}}{R_{1}^{3}C^{2}}} \quad \implies \quad T(S) = \frac{\frac{2}{RC}S}{S^{2}+\frac{1}{RC}S+\frac{1}{R_{1}^{2}C^{2}}} \\ \\
K\frac{\omega_{0}}{Q} &= \frac{2}{RC} \quad y \quad \frac{\omega_{0}}{Q} = \frac{1}{RC} \quad y \quad \omega_{0}^{2} = \frac{1}{R_{1}^{2}C^{2}} \quad Para\,simular\,la\,transferencia\,normalizada\,se\,toma:\, \omega_{0} = 1 \\ \\
K\frac{1}{Q} &= \frac{2}{RC} \quad y \quad \frac{1}{Q} = \frac{1}{RC} \quad y \quad 1 = \frac{1}{R_{1}^{2}C^{2}} \quad \implies \quad \boxed{K = 2} \quad ; \quad R = R_{1}Q \quad y \quad Q = 20 \quad \implies \quad T(S) = \frac{K\frac{1}{Q}S}{S^{2}+\frac{1}{Q}S+1}
\end{align*}$$

```python
#Librerias Externas
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
import numpy as np

#Libreria TC2
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot

w0 = 1
Q = 20
K = 2

TF = TransferFunction([K/Q,0], [1,1/Q,1])

plt.close('all')

bodePlot(TF, fig_id=1, filter_description= 'Q={}'.format(Q))

pzmap(TF, fig_id=2, filter_description= 'Q={}'.format(Q))

GroupDelay(TF, fig_id=3, filter_description= 'Q={}'.format(Q))
```
