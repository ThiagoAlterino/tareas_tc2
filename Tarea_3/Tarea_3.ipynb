## 1) Metodo Storch

$$\begin{align*}
T(S) &= \frac{1}{cosh(S) + senh(S)} \quad y \quad cotgh(S) = \frac{cosh(S)}{senh(S)} = \frac{1}{S} + \frac{1}{\frac{3}{S}+\frac{1}{\frac{5}{S}+\frac{1}{\frac{7}{S}+...}}}
\end{align*}$$

Para n = 2:
$$\begin{align*}
cotgh(S) &= \frac{1}{S} + \frac{S}{3} = \frac{3+S^{2}}{3S} \quad \implies \quad \boxed{T_{2}(S) = \frac{3}{S^{2}+3S+3}}
\end{align*}$$

Para n = 3:
$$\begin{align*}
cotgh(S) &= \frac{1}{S} + \frac{1}{\frac{3}{S}+\frac{S}{5}} = \frac{1}{S} + \frac{1}{\frac{15+S^{2}}{5S}} = \frac{1}{S} + \frac{5S}{15+S^{2}} = \frac{S^{2}+15+5S^{2}}{S(S^{2}+15)} \quad \implies \quad cotgh(S) = \frac{6S^{2}+15}{S^{3}+15S}
\end{align*}$$
$$\boxed{T_{3}(S) = \frac{15}{S^{3}+6S^{2}+15S+15}}$$

Para n = 4:
$$\begin{align*}
cotgh(S) &= \frac{1}{S} + \frac{1}{\frac{3}{S}+\frac{1}{\frac{5}{S}+\frac{S}{7}}} = \frac{1}{S} + \frac{1}{\frac{3}{S}+\frac{7S}{35+S^{2}}} = \frac{1}{S} + \frac{1}{\frac{3(35+S^{2}+7S^{2})}{S(S^{2}+35)}} = \frac{1}{S} + \frac{S^{3}+35S}{3S^{2}+7S^{2}+105} = \frac{10S^{2}+105+S^{4}+35S^{2}}{S(10S^{2}+105)} \\ \\
cotgh(S) &= \frac{S^{4}+45S^{2}+105}{10S^{3}+105S} \quad \implies \quad \boxed{T_{4}(S) = \frac{105}{S^{4}+10S^{3}+45S^{2}+105S+105}}
\end{align*}$$

## 2)Elegir la T(S) con el mínimo orden que cumple con alpha max = 1 dB

$$T(\omega) = T(S)|_{S=j\omega} \quad \quad El\, \,orden\, \,mínimo\, \,es\, \,el\,\,que \left|T_{n}(\omega_{p})\right| = 1 dB $$ 

$$\begin{align*}
T_{2}(\omega) &= \frac{3}{(j\omega)^{2}+3(j\omega)+3} = \frac{3}{-\omega^{2}+j3\omega+3} = \frac{3}{3-\omega^{2} + j 3\omega} \quad \implies \quad T_{2}(\omega) = \frac{3}{\sqrt{(3-\omega^{2})^{2}+9\omega^{2}}}e^{-arctg(\frac{3\omega}{3-\omega^{2}})} \\ \\
\left| T_{2}(\omega) \right| &= \frac{3}{\sqrt{(3-\omega^{2})^{2}+9\omega^{2}}} \quad \implies \quad \left| T_{2}(\omega_{p}) \right| = \frac{3}{\sqrt{(3-\omega^{2}_{p})^{2}+9\omega^{2}_{p}}} = \frac{3}{\sqrt{(3-1^{2})^{2}+9·1^{2}}} = \frac{3}{\sqrt{2^{2}+9}}= \frac{3}{\sqrt{13}}
\end{align*}$$
$$\boxed{\left| T_{2}(\omega_{p}) \right|= \frac{3}{\sqrt{13}} = 0,832}$$

$$\begin{align*}
T_{3}(\omega) &= \frac{15}{(j\omega)^{3}+6(j\omega)^{2}+15(j\omega)+15} = \frac{15}{-j\omega^{3}-6\omega^{2}+j15\omega+15} = \frac{15}{15-6\omega^{2} + j(15\omega-\omega^{3})} \\ \\
T_{3}(\omega) &= \frac{15}{\sqrt{(15-6\omega^{2})^{2}+(15\omega-\omega^{3})^{2}}}e^{-arctg(\frac{15\omega-\omega^{3}}{15-6\omega^{2}})} \quad \implies \quad \left| T_{3}(\omega) \right| = \frac{15}{\sqrt{(15-6\omega^{2})^{2}+(15\omega-\omega^{3})^{2}}} \\ \\
\left| T_{3}(\omega_{p}) \right| &= \frac{15}{\sqrt{(15-6\omega_{p}^{2})^{2}+(15\omega_{p}-\omega_{p}^{3})^{2}}} = \frac{15}{\sqrt{(15-6·1^{2})^{2}+(15·1-1^{3})^{2}}} = \frac{15}{\sqrt{(15-6)^{2}+(15-1)^{2}}}
\end{align*}$$
$$\boxed{\left| T_{3}(\omega_{p}) \right| = \frac{15}{\sqrt{81+196}} = \frac{15}{\sqrt{277}} = 0,9}$$

$$\begin{align*}
T_{4}(S) &= \frac{105}{(j\omega)^{4}+10(j\omega)^{3}+45(j\omega)^{2}+105(j\omega)+105} = \frac{105}{\omega^{4}-45\omega^{2}+105+j(105\omega-10\omega^{3})} \\ \\
T_{4}(S) &= \frac{105}{\sqrt{(\omega^{4}-45\omega^{2}+105)^{2}+(105\omega-10\omega^{3})^{2}}}e^{-arctg(\frac{105\omega-10\omega^{3}}{\omega^{4}-45\omega^{2}+105})} \quad \implies \quad \left| T_{4}(\omega) \right| = \frac{105}{\sqrt{(\omega^{4}-45\omega^{2}+105)^{2}+(105\omega-10\omega^{3})^{2}}}
\end{align*}$$
$$\boxed{\left| T_{4}(\omega_{p}) \right| = \frac{105}{\sqrt{(1-45+105)^{2}+(105-10)^{2}}} = 0,93}$$
