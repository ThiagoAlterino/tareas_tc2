## 2)Elegir la T(S) con el mínimo orden que cumple con alpha max = 1 dB

## 3) Evaluar el Retardo de Grupo D(2,5). Expresar en forma porcentual [%] el error o desviamiento respecto a D(0)

El retardo de grupo es igual a: $\quad D(\omega) = -\frac{\partial\phi(\omega)}{\partial \omega}$
$$T_{3}(\omega) = \frac{15}{\sqrt{(15-6\omega^{2})^{2}+(15\omega-\omega^{3})^{2}}}e^{-arctg(\frac{15\omega-\omega^{3}}{15-6\omega^{2}})} \quad \implies \quad \phi(\omega) = -arctg(\frac{15\omega-\omega^{3}}{15-6\omega^{2}})$$

$$D(\omega) = -\frac{\partial}{\partial \omega} (-arctg(\frac{15\omega-\omega^{3}}{15-6\omega^{2}})) \quad \implies \quad D(\omega) = \frac{6\omega^{4} + 45\omega^{2} + 225}{\omega^{6} + 6\omega^{4} + 45\omega^{2} + 225} \quad \implies \quad D(0) = 1 \quad y \quad D(2,5) = 0,752$$
$$Error \,\, \% = \frac{D(0) - D(2,5)}{D(0)} 100\% = \frac{1-0,752}{1}100\% = 24,8\% \quad \implies \quad \boxed{Error \,\, \% = 24,8\%}$$

## 4) Sintetizar el circuito normalizado con estructuras Sallen-Key con K=1
$$T_{3}(S) = \frac{15}{S^{3}+6S^{2}+15S+15} = \frac{6,459}{S^{2} + 3,678 S + 6,459}\frac{2,322}{S+2,322}$$

Para sintetizar este circuito se cascadea un circuito Sallen-Key de segundo orden y un filtro RC de primer orden. La transferencia de un circuito Sallen-Key con K = 1 es:
$$\boxed{H(S) = \frac{\frac{1}{R_{1}R_{2}C_{1}C_{2}}}{S^{2} + (\frac{1}{R_{1}C_{1}}+\frac{1}{R_{2}C_{1}})S + \frac{1}{R_{1}R_{2}C_{1}C_{2}}}}$$

Entonces:
$$\frac{1}{R_{1}R_{2}C_{1}C_{2}} = 6,459 \quad y \quad \frac{1}{R_{1}C_{1}}+\frac{1}{R_{2}C_{1}} = 3,678$$

Se adopta $\,\, R_{1} = R_{2} = 1$, entonces:
$$\frac{1}{C_{1}C_{2}} = 6,459 \quad y \quad \frac{1}{C_{1}}+\frac{1}{C_{1}} = 3,678 \quad \implies \quad \frac{2}{C_{1}} = 3,678 \quad \implies \quad \boxed{C_{1} = \frac{2}{3,678} = 0,544} \quad \implies \quad \boxed{C_{2} = \frac{1}{6,459 C_{1}} = 0,285}$$

Para la etapa de primer orden se sintetizara un filtro RC cuya transferencia es: $\quad H(S) = \frac{\frac{1}{R_{3}C_{3}}}{S + \frac{1}{R_{3}C_{3}}}$, entonces:

$$\frac{1}{R_{3}C_{3}} = 2,322 \quad se adopta R_{3} = 1 \quad \implies \quad \frac{1}{C_{3}} = 2,322 \quad \implies \quad \boxed{C_{3} = \frac{1}{2,322} = 0,43}$$

## Simulación numérica en python utilizando la función pretty_print_lti

```python
#Librerias Externas
from scipy import signal

#Libreria TC2
from pytc2.sistemas_lineales import pretty_print_lti

#  Bessel-Thompson de orden 2, 3 y 4

# Bessel-Thompson de orden 2
z2,p2 = signal.bessel(2, Wn=1, btype='low', analog=True, norm='delay')
print("Bessel-Thompson de orden 2:")
pretty_print_lti(z2,p2)

# Bessel-Thompson de orden 3
z3,p3 = signal.bessel(3, Wn=1, btype='low', analog=True, norm='delay')
print("Bessel-Thompson de orden 3:")
pretty_print_lti(z3,p3)

# Bessel-Thompson de orden 4
z4,p4 = signal.bessel(4, Wn=1, btype='low', analog=True, norm='delay')
print("Bessel-Thompson de orden 4:")
pretty_print_lti(z4,p4)
```
