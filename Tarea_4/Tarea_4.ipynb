### 2) Obtener la función transferencia

$$\begin{align*}
|T(S)|·|T(-S)| &= \frac{1}{1+\xi^2(C_3(\omega))^2} \quad y \quad C_3(\omega)) = 4\omega^3-3\omega \quad \implies \quad |T(S)|·|T(-S)| = \frac{1}{1+\xi^2(4\omega^3-3\omega)^2} \\ \\
|T(S)|·|T(-S)| &= \frac{1}{1+\xi^2(16\omega^6-24\omega^4+9\omega^2)} \quad \implies \quad |T(S)|·|T(-S)| = \frac{\frac{1}{16\xi^2}}{\omega^6-\frac{24}{16}\omega^4+\frac{9}{16}\omega^2+\frac{1}{16\xi^2}} \\ \\
\omega &= \frac{S}{j} \quad \implies \quad \omega^6 = -S^6 \quad y \quad \omega^4 = S^4 \quad y \quad \omega^2 = -S^2 \\ \\
|T(S)|·|T(-S)| &= \frac{\frac{1}{16\xi^2}}{-S^6-\frac{24}{16}S^4-\frac{9}{16}S^2+\frac{1}{16\xi^2}} \quad \implies \quad |T(S)|·|T(-S)| = \frac{-\frac{1}{16\xi^2}}{S^6+\frac{24}{16}S^4+\frac{9}{16}S^2-\frac{1}{16\xi^2}} \\ \\
|T(S)|·|T(-S)| &= \frac{-\frac{1}{16\xi^2}}{(S+0.626)(S^2+0.626·S+1.142)(S^2-0.626·S+1.142)(S-0.626)} \\ \\
T(S) &= \frac{\sqrt{\frac{1}{16\xi^2}}}{(S+0.626)(S^2+0.626·S+1.142)} \quad \implies \quad T(S) = \frac{0.715}{(S+0.626)(S^2+0.626·S+1.142)}
\end{align*}$$
$$\boxed{T(S) = \frac{0.626}{(S+0.626)}\frac{1.142}{(S^2+0.626·S+1.142)}}$$

### 3) Aplicar el nucleo de transformada pasabanda

$$T(S) = \frac{0.715}{S^3 + 1.252 S^2 + 1.534 S + 0.715} \quad y\,\, el\,\, nucleo\,\, de\,\, transformada\,\, pasa\,\, banda: \quad S = \frac{$^2 + \omega_0^2}{\frac{\omega_0}{Q}$}$$
$$T($) = \frac{0.715}{(\frac{$^2 + \omega_0^2}{\frac{\omega_0}{Q}$})^3+1.252(\frac{$^2 + \omega_0^2}{\frac{\omega_0}{Q}$})^2+1.534(\frac{$^2 + \omega_0^2}{\frac{\omega_0}{Q}$})+0.715} = \frac{0.715}{\frac{($^2 + \omega_0^2)^3}{(\frac{\omega_0}{Q}$)^3}+1.252\frac{($^2 + \omega_0^2)^2}{(\frac{\omega_0}{Q}$)^2}+1.534\frac{$^2 + \omega_0^2}{\frac{\omega_0}{Q}$}+0.715}$$
$$T($) = \frac{0.715(\frac{\omega_0}{Q}$)^3}{($^2+\omega_0^2)^3 + 1.252($^2+\omega_0^2)^2(\frac{\omega_0}{Q}) + 1.534($^2+\omega_0^2)(\frac{\omega_0}{Q})^2+0.715(\frac{\omega_0}{Q})^3} \quad se\,\, toma\,\, \omega\,\, normalizada \quad \implies \quad \omega = 1 \quad y\,\, Q = 5\,\, es\,\, dato$$
$$\begin{align*}
T($) &= \frac{5.72x10^{-3}$^3}{$^6 + 0.25 $^5 + 3.06 $^4 + 0.506 $^3 + 3.06 $^2 + 0.25 $ + 1} \quad \implies \quad T($) = \frac{5.72x10^{-3}$^3}{($^2+0.125 $ + 1)($^2+0.069 $ + 1.226)($^2+0.056 $ + 0.815)} \\ \\

### 4) Simulación numérica en python del filtro pasa banda total y las de las SOS

#### Simulación del filtro pasa banda total:

$$\boxed{T(S) = \frac{5.72x10^{-3}S^3}{(S^2+0.125 S + 1)(S^2+0.069 S + 1.226)(S^2+0.056 S + 0.815)}}$$

```python
#Librerias Externas
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np

#Libreria TC2
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot

#  Implementacion de filtro pasa banda Chebyshev de orden 6

# 1er etapa filtro pasabajo de orden 2
num1 = [1,0]
den1 = [1,0.125,1]

# 2da etapa filtro pasabajo de orden 2
num2 = [1,0]
den2 = [1,0.069,1.226]

# 3er etapa filtro pasabajo de orden 2
num3 = [0.00572,0]
den3 = [1,0.056,0.815]

num = np.convolve(num1, np.convolve(num2, num3))
den = np.convolve(den1, np.convolve(den2, den3))

TF = signal.TransferFunction(num, den)

plt.close('all')

bodePlot(TF, fig_id=1)

pzmap(TF, fig_id=2)

GroupDelay(TF, fig_id=3)
```

### Simulación Circuital del filtro pasa banda implementado mediante secciones pasivas separadas por seguidores de tensión activos. 

El filtro en la banda de paso deberia llegar a los 0dB como se ve en la simulacion numerica, pero no supe como implementar la constante k = 11.84

$$T(S) = 11.84·\frac{0.125 S}{S^2+0.125 S + 1}·\frac{0.069 S}{S^2+0.069 S + 1.226}·\frac{0.815 S}{S^2+0.056 S + 0.815}$$

Para la implementacion de las tres etapas se utilizara el siguiente circuito pasivo:
![circuito](./archivos/circuito.png)

Primer etapa:
$$T(S) = \frac{0.125 S}{S^2+0.125 S + 1} \quad \implies \quad \frac{R_1}{L_1} = 0.125 \quad y \quad \frac{1}{L_1C_1} = 1 \quad se\,\, adopta\,\, C_1 = 1 \quad \implies \quad \frac{1}{L_1} = 1 \quad \implies \quad \boxed{L_1 = 1 \quad y \quad R_1 = 0.125}$$

Segunda etapa:
$$T(S) = \frac{0.069 S}{S^2+0.069 S + 1.226} \quad \implies \quad \frac{R_2}{L_2} = 0.069 \quad y \quad \frac{1}{L_2C_2} = 1.226 \quad se\,\, adopta\,\, C_2 = 1 \quad \implies \quad \frac{1}{L_2} = 1.226 \quad \implies \boxed{L_2 = 0.816 \quad y \quad R_2 = 0.056}$$

$$T(S) = \frac{0.815 S}{S^2+0.056 S + 0.815} \quad \implies \quad \frac{R_3}{L_3} = 0.056 \quad y \quad \frac{1}{L_3C_3} = 0.815 \quad se\,\, adopta\,\, C_3 = 1 \quad \implies \quad \frac{1}{L_3} = 0.815 \quad \implies \boxed{L_3 = 1.227 \quad y \quad R_1 = 0.069}$$
