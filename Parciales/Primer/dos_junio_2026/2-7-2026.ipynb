## 1)A) Filtro pasa alto maxima planicidad transferencia normalizada

$$\alpha_{max} = 0.5 dB \quad ; \quad f_C = 20KH_Z \quad ; \quad \alpha_{min} = 20dB \,\, en \,\, f_S = 5KH_Z$$

Para calcular el filtro pasa alto primero hay que normalizar las frecuencias de corte y de paso, transformarlas a sus equivalente para un filtro pasa bajo y con las frecuencias para el filtro pasa bajo encontrar el orden y la transferencia pasa bajo.
Una vez tenga la transferencia del filtro pasa bajo aplicar la propiedad de transformacion pasa alto y ahi obtendriamos la transferencia pasa alto.

Primero se normalizan las frecuencias:
$$F_S = \frac{f_S}{f_C} = \frac{5KH_Z}{20KH_Z} = 0.25 \quad y \quad F_C = \frac{f_C}{f_C} = 1$$
Para realizar la transformacion en frecuencia para calcular la transferencia pasa bajo hay que realizar: 
$$\Omega = \frac{1}{\omega} \quad \implies \quad \Omega_C = 1 \quad y \quad \Omega_S = 4$$

$$\xi² = 10^{\frac{\alpha_{max}}{10}}-1 \quad \implies \quad \xi² = 0.122 \quad y \quad \alpha_{min} = 10 Log(1+\xi² · \Omega_S^{2n})$$

| n | $\alpha_{n}$ |
| :--- | :--- |
| 2 | 15 dB |
| 3 | 27 dB |
| 4 | 39 dB |

Para el orden 3 el filtro ya cumple con la exigencia del enunciado de que la atenuacion minima en 5KHz sea de 20dB.

### Transferencia pasa bajo normalizada maxima planicidad de orden 3
$$T(S) = \frac{1}{S+1}\frac{1}{S² + 2·cos(\frac{\pi}{3})·S + 1}$$
Esta es la transferencia Butterworth, lo que implica que $\xi² = 1$, pero  en este caso $\xi² = 0.122$ esto implica que la transferencia de maxima planicidad con $\xi² \neq 1$:
$$T(S) = \frac{R}{S+R}\frac{R²}{S² + 2·cos(\frac{\pi}{3})·R·S + R²} \quad y \quad R = \xi^{-\frac{1}{n}} = 0.349^{-\frac{1}{3}}=1.42 \quad y \quad R² = 2.0162 \quad y \quad 2·cos(\frac{\pi}{3}) = 1 \quad \implies \quad 2·cos(\frac{\pi}{3})·R = 1.42$$

$$\boxed{T(S) = \frac{1.42}{S+1.42}\frac{2.0162}{S² + 1.42·S + 2.0162}}$$

### Transformacion pasa alto
Para aplicar la transformacion pasa alto hay que hacer $T(\$) = T(S)|_{S = \frac{1}{\$}}$, entonces:
$$T(\$) = \frac{1.42}{\frac{1}{\$}+1.42}·\frac{2.0162}{\frac{1}{\$²} + 1.42·\frac{1}{\$} + 2.0162} = \frac{1.42}{\frac{1}{\$}(1+1.42·\$)}·\frac{2.0162}{\frac{1}{\$²}(1 + 1.42·\$ + 2.0162·\$²)} = \frac{1.42·\$}{1.42·\$ + 1}·\frac{2.0162·\$²}{2.0162·\$²+ 1.42·\$ + 1}$$
$$T(\$) = \frac{\$}{\$ + \frac{1}{1.42}}·\frac{\$²}{\$²+ \frac{1.42}{2.0162}·\$ + \frac{1}{2.0162}} \quad \implies \quad T(\$) = \frac{\$}{\$ + 0.704}·\frac{\$²}{\$²+ 0.704·\$ + 0.496}$$

### Transferencia pasa alto normalizada maxima planicidad de orden 3
$$\boxed{T(S) = \frac{S}{S + 0.704}·\frac{S²}{S²+ 0.704·S + 0.496}}$$
