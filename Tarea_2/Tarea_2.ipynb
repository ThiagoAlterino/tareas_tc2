## 1) Analizar la impedancia de entrada desde el nodo Vx. Hallar los valores de R4,R5,Y1,Y2 e Y3 de tal manera que responda como un Inductor de valor unitario.

![analisiscircuito](./archivos/analisiscircuito.png)

$$\begin{align*}
I_{R1} = I_{C} + I_{R2} + I_{R3-1} \quad y \quad I_{R1} = \frac{V_{1}-0}{R_{1}} = \frac{V_{1}}{R_{1}} \quad ; \quad I_{C} = (0-V_{x})SC = -SCV_{x} \quad ; \quad I_{R2} = \frac{0-V_{x}}{R_{2}} = -\frac{V_{x}}{R_{2}} \quad ; \quad I_{R3-1} = \frac{0-V_{2}}{R_{3}} = -\frac{V_{2}}{R_{3}}
\end{align*}$$

$$\begin{align*}
I_{3-2} = \frac{V_{x}-0}{R_{3}} = (0-(-V_{2}))SC \quad \implies \quad \frac{V_{x}}{R_{3}} = SCV_{2} \quad \implies \quad .V_{x} = SCR_{3}V_{2}
\end{align*}$$

$$\begin{align*}
I_{R1} &= I_{C} + I_{R2} + I_{R3-1} \quad \implies \quad \frac{V_{1}}{R_{1}} = -SCV_{x} - \frac{V_{x}}{R_{2}} - \frac{V_{2}}{R_{3}}\quad \implies \quad \frac{V_{1}}{R_{1}} = -(SC + \frac{1}{R_{2}})V_{x} -\frac{V_{2}}{R_{3}} \\ \\
\frac{V_{1}}{R_{1}} &= -(\frac{SCR_{2}+1}{R_{2}})SCR_{3}V_{2} -\frac{V_{2}}{R_{3}} \quad \implies \quad \frac{V_{1}}{R_{1}} = -(\frac{S^{2}C^{2}R_{2}R_{3}+SCR_{3}}{R_{2}} +\frac{1}{R_{3}})V_{2} \\ \\
\frac{V_{1}}{R_{1}} &= -(\frac{S^{2}C^{2}R_{2}R_{3}^{2}+SCR_{3}^{2}+R_{2}}{R_{2}R_{3}})V_{2} \quad \implies \quad T(S) = \frac{V_{2}}{V_{1}} = -\frac{R_{2}R_{3}}{R_{1}(S^{2}C^{2}R_{2}R_{3}^{2}+SCR_{3}^{2}+R_{2})} \\ \\
T(S) &= -\frac{R_{2}R_{3}}{R_{1}R_{2}R_{3}^{2}C^{2}(S^{2}+S\frac{R_{3}^{2}C}{R_{2}R_{3}^{2}C^{2}} + \frac{R_{2}}{R_{2}R_{3}^{2}C^{2}})} \quad \implies \quad T(S) = -\frac{\frac{1}{R_{1}R_{3}C^{2}}}{S^{2}+\frac{1}{R_{2}C}S+\frac{1}{R_{3}^{2}C^{2}}} \\ \\
\omega_{0}^{2}&=\frac{1}{R_{3}^{2}C^{2}} \quad ; \quad \frac{\omega_{0}}{Q} = \frac{1}{R_{2}C} \quad ; \quad K\omega_{0}^{2} = -\frac{1}{R_{1}R_{3}C^{2}} \quad \implies \quad \omega_{0} = \frac{1}{R_{3}C} \quad y \quad Q = \frac{R_{2}}{R_{3}} \quad y \quad K = -\frac{R_{3}}{R_{1}} \\ \\
T(S) &= \frac{K\omega_{0}^{2}}{S^{2}+\frac{\omega_{0}}{Q}S+\omega_{0}^{2}} \quad \implies \quad \boxed{T(S) = -\frac{\frac{R_{3}}{R_{1}}\frac{1}{R_{3}^{2}C^{2}}}{S^{2}+\frac{1}{R_{3}C}\frac{R_{3}}{R_{2}}S+\frac{1}{R_{3}^{2}C^{2}}}}
