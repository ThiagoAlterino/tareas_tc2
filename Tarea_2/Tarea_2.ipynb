## 2) Implementar un filtro pasabajos Chebyshev que cumpla con la siguiente plantilla. Debera interconectar mas de una red como la aquí presentada en configuración cascada. De ser necesario incluya una red de orden 1 pasabajos RC.

Para implementar el filtro Chebyshev primero hay que determinar el orden del filtro para esto utilizando las formulas $ \xi^{2} = 10^{\frac{\alpha_{max}}{10}}-1$ y $\alpha_{n} = 10Log(1+\xi^{2}cosh^{2}(n·cosh^{-1}(\omega_{s})))dB$ se itera hasta encontrar una atenuacion $\alpha_{n}$ igual o mayor a $\alpha_{min}$

| n | $\alpha_{n}$ |
| :--- | :--- |
| 2 | 14,6 dB |
| 3 | 29,76 dB |
| 4 | 45,07 dB |
| 5 | 60,38 dB |

El filtro pasabajo Chebyshev debera ser de orden 5, entonces:
$$\begin{align*}
\left| T(S) \right|·\left| T(-S) \right| &= \frac{1}{1+\xi^{2}[C_{5}(\omega)]^{2}} \quad y \quad C_{n} = 2\omega C_{n-1} - C_{n-2} \\ \\
C_{0} &= 1 \quad ; \quad C_{1} = \omega \quad ; \quad C_{2} = 2\omega^{2} - 1 \quad ; \quad C_{3} = 4\omega^{3} - 2\omega - \omega \quad ; \quad C_{4} = 8\omega^{4} - 8\omega^{2} + 1 \quad ; \quad C_{5} = 16\omega^{5} - 20\omega^{3} + 5\omega \\ \\
\left| T(S) \right|·\left| T(-S) \right| &= \frac{1}{1+\xi^{2}(16\omega^{5} - 20\omega^{3} + 5\omega)^{2}} \quad \implies \quad \left| T(S) \right|·\left| T(-S) \right| = \frac{1}{1+\xi^{2}(256\omega^{10}-640\omega^{8}+560\omega^{6}-200\omega^{4}+25\omega^{2})} \\ \\
\left| T(S) \right|·\left| T(-S) \right| &= \frac{\frac{1}{256\xi^{2}}}{\frac{1}{256\xi^{2}}+\omega^{10}-\frac{640}{256}\omega^{8}+\frac{560}{256}\omega^{6}-\frac{200}{256}\omega^{4}+\frac{25}{256}\omega^{2}} \quad y \quad \omega = \frac{S}{j}  \implies  (\frac{S}{J})^{2} = -S^{2} \,;\, (\frac{S}{J})^{4} = S^{4} \,;\, (\frac{S}{J})^{6} = -S^{6} \,;\, (\frac{S}{J})^{8} = S^{8} \,;\, (\frac{S}{J})^{10} = -S^{10} \\ \\
\left| T(S) \right|·\left| T(-S) \right| &= \frac{\frac{1}{256\xi^{2}}}{-(S^{10}+\frac{640}{256}S^{8}+\frac{560}{256}S^{6}+\frac{200}{256}S^{4}+\frac{25}{256}S^{2}-\frac{1}{256\xi^{2}})} \\ \\
\left| T(S) \right|·\left| T(-S) \right| &= \frac{-40,49x10^{-3}}{(S+0,386)(S^{2}+0,625 S+0,495)(S^{2}+0,239 S+1,054)(S-0,386)(S^{2}-0,625 S+0,495)(S^{2}-0,239 S+1,054)} \\ \\
T(S) &= \frac{201x10^{-3}}{(S+0,386)(S^{2}+0,625 S+0,495)(S^{2}+0,239 S+1,054)} \quad \implies \quad \boxed{T(S) = \frac{0,495}{S^{2}+0,625 S+0,495}\frac{1,054}{S^{2}+0,239 S+1,054}\frac{0,386}{S+0,386}}
\end{align*}$$

Para implementar este filtro de orden 5 se van a interconectar en cascada dos filtros pasabajo del punto 1) y un filtro pasabajo RC

Las dos primeras redes fueron analizadas en el punto anterior, su transferencia es:

$$\begin{align*}
T(S) &= -\frac{K\omega_{0}^{2}}{S^{2}+\frac{\omega_{0}}{Q}S+\omega_{0}^{2}} \quad y \quad \omega_{0} = \frac{1}{R_{3}C} \quad y \quad Q = \frac{R_{2}}{R_{3}} \quad y \quad K = \frac{R_{3}}{R_{1}}
\end{align*}$$

Red 1: $\quad T(S) = \frac{0,495}{S^{2}+0,625 S+0,495}$
$$\begin{align*}
K\omega_{0}^{2} &= 0,495 \quad y \quad \omega_{0}^{2} = 0,495 \quad \implies \quad K = 1 \quad \implies \quad \boxed{R_{1} = R_{3}} \quad y \quad \boxed{\frac{1}{R_{3}C} = \sqrt{0,495}} \\ \\
\frac{\omega_{0}}{Q} &= 0,625 \quad \implies \quad \frac{1}{R_{3}C}\frac{R_{3}}{R_{2}} = 0,625 \quad \implies \quad \boxed{\frac{1}{R_{2}C} = 0,625}
\end{align*}$$

Se adopta $C = 1$ y $R_{4} = R_{3}$, entonces:
$$\begin{align*}
\frac{1}{R_{3}} = \sqrt{0,495} \quad \implies \quad \boxed{R_{3} = \frac{1}{\sqrt{0,495}} = 1,421} \quad \implies \quad \boxed{R_{1} = R_{4} = 1,421} \quad y \quad \frac{1}{R_{2}} = 0,625 \quad \implies \quad \boxed{R_{2} = \frac{1}{0,625} = 1,6}
\end{align*}$$

Red 2:  $\quad T(S) = \frac{1,054}{S^{2}+0,239 S+1,054}$
$$\begin{align*}
K\omega_{0}^{2} &= 1,054 \quad y \quad \omega_{0}^{2} = 1,054 \quad \implies \quad K = 1 \quad \implies \quad \boxed{R_{1} = R_{3}} \quad y \quad \boxed{\frac{1}{R_{3}C} = \sqrt{1,054}} \\ \\
\frac{\omega_{0}}{Q} &= 0,239 \quad \implies \quad \frac{1}{R_{3}C}\frac{R_{3}}{R_{2}} = 0,239 \quad \implies \quad \boxed{\frac{1}{R_{2}C} = 0,239}
\end{align*}$$

Se adopta $C = 1$ y $R_{4} = R_{3}$, entonces:
$$\begin{align*}
\frac{1}{R_{3}} = \sqrt{1,054} \quad \implies \quad \boxed{R_{3} = \frac{1}{\sqrt{1,054}} = 0,974} \quad \implies \quad \boxed{R_{1} = R_{4} = 0,974} \quad y \quad \frac{1}{R_{2}} = 0,239 \quad \implies \quad \boxed{R_{2} = \frac{1}{0,239} = 4,184}
\end{align*}$$

Red 3 (filtro RC): $\quad T(S) = \frac{0,386}{S+0,386}$

Transferencia de un filtro RC:  $$T(S) = \frac{\frac{1}{RC}}{S+\frac{1}{RC}} \quad \implies \quad \boxed{\frac{1}{RC} = 0,386}$$

Se adopta $C = 1$, entonces:
$$\frac{1}{R} = 0,386 \quad \implies \quad \boxed{R = \frac{1}{0,386} = 2,59}$$

```python
#Librerias Externas
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np

#Libreria TC2
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot

#  Implementacion de filtro pasabajo Chebyshev de orden 5 cascadeando dos pasabajo de orden 2 y un filtro RC

# 1er etapa filtro pasabajo de orden 2
num1 = [0.495]
den1 = [1,0.625,0.495]

# 2da etapa filtro pasabajo de orden 2
num2 = [1.054]
den2 = [1,0.239,1.054]

# 3er etapa filtro pasabajo de orden 2
num3 = [0.386]
den3 = [1,0.386]

num = np.convolve(num1, np.convolve(num2, num3))
den = np.convolve(den1, np.convolve(den2, den3))

TF = signal.TransferFunction(num, den)

plt.close('all')

bodePlot(TF, fig_id=1)

pzmap(TF, fig_id=2)

GroupDelay(TF, fig_id=3)
```
