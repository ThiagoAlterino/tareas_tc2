## Primer diseño
Para resolver este problema primero planteamos:

$$\Omega_{\omega} = \frac{D_{n}(0)}{D(0)} \frac{rad}{s} \quad y \quad D(0) = 100 \micro s \quad \implies \quad \Omega_{\omega} = \frac{1 s}{100 \micro s} \frac{rad}{s} = 10 K \frac{rad}{s} \quad \implies \quad \boxed{\Omega_{\omega} = 10 K \frac{rad}{s}}$$

El problema que encontrabamos aca es que al normalizar el $\omega_p = 2·\pi·5KH_z$ para calcular calcular la atenuacion y determinar cual orden deberiamos tomar para implementar el filtro. Recien en un orden 22 la atenuacion es 1dB(lo calculamos con un codigo).

Si nosotros tomamos la transferencia de orden tres y calculamos el circuito nos da que cumple con el retardo de grupo, pero el desvio maximo en 3KHz y la atenuacion maxima en 5KHz no se cumplen.

## Segundo diseño

Como en la primera resolucion no llegamos a nada, planteamos lo siguiente:
![resolucion1](./archivos/resolucion1.png)

Luego calculamos la transferencia de los Bessel de distintos ordenes, calculamos en $\Omega_p$ y tomamos la transferencia del orden 3 cuya atenuacion en $\Omega_p$ es la que mejor aproxima al $\alpha_{MAX} = 1dB$

En este caso el problema que encontramos es que no se cumple el retardo de grupo, ya que nos da 32 $\micro s$. Pero si se cumple el desvio maximo en 3KHz y la atenuacion maxima en 5Hz
